## Data Used: https://data.ibb.gov.tr/dataset/istanbul-genel-yesil-alan-bilgiler-2004-2019/resource/d6527f7d-07b4-4dd9-aa09-76e92d86526a

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
plt.style.use("ggplot")
```

## Importing the selected dataset into python with pandas and converting the dataframe

```python
# data_path = path to data file
data_path="/home/berkay/YapayZekaKampı-ZeminIstanbul/DataVisualizationAssignment/istanbul-genel-yeil-alan-bilgiler-2004-2019.xlsx"
sheetName="Yeni Yapılan Yeşil Alan Mik."
data=pd.read_excel(data_path,sheet_name=sheetName,index_col="FAALİYET KONUSU")
green_area=pd.DataFrame(data)
green_area=green_area.drop(columns="TOPLAM",axis=1)
green_area
```

## Shows the row-column size of the dataset

```python
green_area.shape
```

## Displaying the data types of the columns in the data set

```python
green_area.dtypes
```

### For an easier analysis, let's first swap the rows and columns, that is, let's apply the transpose operation.

```python
green_area=green_area.transpose()
green_area.head(3)
```

## Checking if there is any missing value in the data set,

```python
green_area.isnull().any()
```

## Any missing value transactions

```python
# Let's fill the nan values in the 'Onarılan Anıt Ağaç Sayısı (Adet)' column with the average value
mean=green_area["Onarılan Anıt Ağaç Sayısı (Adet)"].mean()
green_area["Onarılan Anıt Ağaç Sayısı (Adet)"]=green_area["Onarılan Anıt Ağaç Sayısı (Adet)"].fillna(mean)
green_area["Onarılan Anıt Ağaç Sayısı (Adet)"].isnull()
```

```python
# Let's check for NAN value one last time.
green_area.isnull().any()
```

## Performing a value conversion to an existing column in the dataset

```python
green_area.head(3)
```

### Change the data type to 'int64' in columns other than "Onarılan Anıt Ağaç Sayısı (Adet)","Yeni Yapılan Yeşil Alan (m²)","Revize Edilen Yeşil Alan (m²)".

```python
columns_name=green_area.columns.to_list()
pop_list=1,1,4
for x in pop_list:
    columns_name.pop(x)
for i in columns_name:
    green_area[i]=green_area[i].astype("int64")
```

## Deriving a new column from existing columns in the dataset

```python
green_area.dtypes
```

### Let's combine the 'Revize Edilen Yeşil Alan (m²)' ,''Bakımı Yapılan Alan (m²)' columns to form the'Yeşil Alanla İlgili Yapılan Toplam Düzenlemeler (m²)' column.


```python
new_Col_Name="Yeşil Alanla İlgili Yapılan Toplam Düzenlemeler (m²)"
green_area[new_Col_Name]=green_area["Revize Edilen Yeşil Alan (m²)"]+green_area["Bakımı Yapılan Alan (m²)"]
green_area.head(3)
```

## Checking the correlations of the columns in the data set

```python
green_area.corr(method="pearson")
# A few reminders about the Pearson correlation;
# When the value of the Pearson correlation is 0, it means that there is no linear relationship between the two variables.
# However, there may be a non-linear relationship between the variables. Therefore, a value of 0 means that the two variables differ from each other.
# does not mean completely independent.
# Pearson correlation values between -1 and 1.
# The Pearson correlation changes when we transform the data.
# The correlation of a variable with itself is 1. Therefore, all diagonal values are 1.00.
```

```python
green_area.corr(method="spearman")
# A few reminders about sperm correlation;
# We can calculate the Spearman rank correlation measure to find the relationship of ranked variables.
#1 for perfectly positively correlated variables
# -1 for perfectly negatively correlated variables
# 0 means there is no correlation between variables.
```

```python
green_area.corr(method="kendall")
# A few reminders about kendall correlation;
# Measures the agreement between two rankings.
# The Kendall correlation measure is known as the Kendall Tau measure.
# It is a non-parametric measure that does not require any assumptions about the common probability distributions of the variables.
```

```python
value =green_area["Bakımı Yapılan Ağaç Sayısı (Adet)"]
step_value=90000

plt.figure(figsize=(12,7))

plt.plot(green_area.index,green_area["Dikilen Ağaç Sayısı (Adet)"],label="Dikilen Ağaç Sayısı (Adet)",color="lightcoral",marker="o")
plt.plot(green_area.index,green_area["Bakımı Yapılan Ağaç Sayısı (Adet)"],label="Bakımı Yapılan Ağaç Sayısı (Adet)",color="mediumpurple",marker="o")

plt.xticks(np.arange(green_area.index[0],green_area.index[15]+1,1))
plt.yticks(np.arange(min(green_area["Bakımı Yapılan Ağaç Sayısı (Adet)"])-10000,max(value)+step_value,step_value))

plt.title("DİKİLEN AĞAÇ SAYISI (ADET) ve BAKIMI YAPILAN AĞAÇ SAYISI (ADET) COMPARISON TABLE ")
plt.xlabel("YEAR")
plt.ylabel("NUMBER OF TREE (NUMBER)")

plt.legend(loc="best")
plt.show()
```

```python
def format_number(data_value, indx):
    if data_value >= 1_000_000:
        formatter = '{:1.1f}M'.format(data_value*0.000_001)
    else:
        formatter = '{:1.0f}K'.format(data_value*0.001)
    return formatter
```

```python
step_value = 250000
values =green_area["Revize Edilen Yeşil Alan (m²)"]

fig, ax = plt.subplots(figsize=(12,7))

ax.barh(green_area.index,values,label="REVİZE EDİLEN MİKTAR(m²)",color="hotpink")

plt.xticks(np.arange(0, max(values)+step_value, step_value))
plt.yticks(np.arange(green_area.index[0],green_area.index[15]+1,1))

ax.xaxis.set_major_formatter(format_number)
ax.set_title("İBB 2004-2019 REVISED GREEN FIELD TABLE")
ax.set_ylabel("YEAR")
ax.set_xlabel("AMOUNT OF GREEN AREA (m²)")
ax.legend(loc="best")

plt.tight_layout()
plt.show()
```

```python
step_value = 500000
values =green_area["Yeni Yapılan Yeşil Alan (m²)"]

fig, ax = plt.subplots(figsize=(12,7))

ax.bar(green_area.index,values,label="NEW GREEN AREA AMOUNT (m²)",color="darkviolet")

plt.yticks(np.arange(0, max(values)+step_value, step_value))
plt.xticks(np.arange(green_area.index[0],green_area.index[15]+1,1))

ax.yaxis.set_major_formatter(format_number)
ax.set_title("İBB 2004-2019 NEW GREEN FIELD TABLE")
ax.set_xlabel("YEAR")
ax.set_ylabel("AMOUNT OF GREEN AREA (m²)")
ax.legend(loc="best")

plt.tight_layout()
plt.show()
```

```python
values=green_area["Yeni Yapılan Yeşil Alan (m²)"]
step_value=400000
w=0.4

fig, ax = plt.subplots(figsize=(12, 7))

bar1=np.arange(len(green_area.index))
bar2=[i+w for i in bar1]
axis_x=np.arange(green_area.index[0],green_area.index[15]+1,1)

ax.bar(bar1,green_area["Yeni Yapılan Yeşil Alan (m²)"],0.4,label="Newly Built Green Area (m²)",color="lightcoral")
ax.bar(bar2,green_area["Revize Edilen Yeşil Alan (m²)"],0.4,label="REVISED GREEN AREA (m²)",color="lightskyblue")

plt.xticks(bar1+w/2,green_area.index)
plt.yticks(np.arange(0, max(values)+step_value, step_value))

ax.yaxis.set_major_formatter(format_number)
ax.set_title("YENİ YAPILAN YEŞİL ALAN (m²) ve REVİZE EDİLEN YEŞİL ALAN (m²) COMPARISON TABLE")
ax.set_xlabel("YEAR")
ax.set_ylabel("AMOUNT OF GREEN AREA (m²)")

plt.tight_layout()
plt.legend(loc="best")
plt.show()
```

```python
x=green_area.index.astype("float64")
y1=green_area["Dikilen Ağaç Sayısı (Adet)"].astype("float64")
y2=green_area["Bakımı Yapılan Ağaç Sayısı (Adet)"].astype("float64")

def format_number(data_value, indx):
    if data_value >= 1_000_000:
        formatter = '{:1.1f}M'.format(data_value*0.000_001)
    else:
        formatter = '{:1.0f}K'.format(data_value*0.001)
    return formatter

fig,ax=plt.subplots(figsize=(12,7))

plt.stackplot(x,y1,y2,labels=["Number of Trees Planted (Number)","Number of Trees Maintained (Number)"],colors=["lightcoral","lightskyblue"])

plt.xticks(np.arange(green_area.index[0],green_area.index[15]+1,1))

ax.yaxis.set_major_formatter(format_number)

plt.title("İBB 2004-2019 AREA TABLE SHOWING THE NUMBER OF TREES PLANT AND THE NUMBER OF TREES MAINTAINED ")
plt.xlabel("YEAR")
plt.ylabel("NUMBER")

plt.tight_layout()
plt.legend(loc="best")
plt.show()
```
